```python
# Import Library
 
import pandas as pd
import numpy as np
import json
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt
import seaborn as sns
```

Get Data by converting

```python

data_list = []

# Read and parse JSON data from each line in the file
with open('trx_sample.txt', 'r') as f:
    for line in f:
        try:
            record = json.loads(line)
            data_list.append(record)
        except json.JSONDecodeError:
            print("Skipping invalid JSON record:", line)

# Create a Pandas DataFrame from the parsed JSON data
df = pd.DataFrame(data_list)
df.head()
```

```python
rows = df.shape[0]  
cols = df.shape[1]  
# Printing the number of Dataframe's rows and columns  
print("Rows: ", rows)  
print("Columns: ", cols)  
```

We have so many nan values but can't be seen without these changes

```python
#Converting blank cells into NaN
df = df.replace(r'^\s*$', np.nan, regex=True)
df.head()
```

```python
#Type of each column in the dataset and null values
df.isnull().sum()
```

```python
df.shape
```

```python
# after dealing with transectiontye column
imputer=SimpleImputer(strategy='most_frequent')
df['transactionType'] = imputer.fit_transform(df[['transactionType']])
```

```python
df.info()
```

```python
# List of columns to convert from 'object' to 'int'
columns_to_convert = ['accountNumber', 'customerId', 'cardCVV', 'enteredCVV', 'cardLast4Digits']

# Convert the specified columns to 'int'
for col in columns_to_convert:
    df[col] = df[col].astype('int')

# Reset the DataFrame index
df = df.reset_index(drop=True)

```

```python
#Extracting each attribute from date time column and creating new column
df['transactionDateTime'] = pd.to_datetime(df['transactionDateTime'], errors='coerce')
df['transaction_year'] = df['transactionDateTime'].dt.year
df['transaction_month'] = df['transactionDateTime'].dt.month
df['transaction_week'] = df['transactionDateTime'].dt.week
df['transaction_day'] = df['transactionDateTime'].dt.day
df['transaction_hour'] = df['transactionDateTime'].dt.hour
df['transaction_minute'] = df['transactionDateTime'].dt.minute

# Reset the DataFrame index
df = df.reset_index(drop=True)
```

```python
print(df['transactionDateTime'])
```

```python
Duplicated = (df.sort_values(['transactionDateTime']).groupby(['accountNumber', 'transactionAmount'], sort=False)['transactionDateTime'].diff().dt.total_seconds().lt(600))
Duplicated.value_counts()
```

```python
df['multiswipe'] = Duplicated
```

```python
# percentage of Multiswipe
multiswipe_count = df['multiswipe'].sum()
total_count = len(df)

multiswipe_percentage = (multiswipe_count / total_count) * 100


```

```python
#Extracting reversal transactions
df['reversal'] = np.where(df['transactionType'] == 'REVERSAL', True,False)
df['reversal'].value_counts()
```

```python
# percentage of reversal
reversed_count = df['reversal'].sum()
total_count = len(df)

reversed_percentage = (reversed_count / total_count) * 100

```

```python
# Describe both types of transactions
if reversed_percentage > 0:
    reversed_description = f"Percentage of reversed transactions: {reversed_percentage:.2f}%"
else:
    reversed_description = "No reversed transactions in the dataset"

if multiswipe_percentage > 0:
    multiswipe_description = f"Percentage of multi-swipe transactions: {multiswipe_percentage:.2f}%"
else:
    multiswipe_description = "No multi-swipe transactions in the dataset"

# Print the descriptions
print(reversed_description)
print(multiswipe_description)
```

Removing Some Colums Means performing cleaning

```python
print("Before dropping columns:", df. columns)
df.drop(['df2','transactionDateTime','customerId','df1', 'echoBuffer','merchantCity', 'merchantState', 'merchantZip', 'posOnPremises', 'recurringAuthInd', 'recurringAuthInd'], axis=1, inplace=True)
print("After dropping columns:", df.columns)
```

```python
df.describe()
```

```python
#Converting boolean values to binary
for col in df.columns:
    if(df[col].astype(str).str.contains('True').sum()>0 and df[col].astype(str).str.contains('False').sum()>0):
        df.loc[df[col].astype(str).str.contains('True'), col] = 1
        df.loc[df[col].astype(str).str.contains('False'), col] = 0
print(df.info)
```

```python
#Converting distinct values using label encoding
from sklearn.preprocessing import LabelEncoder
obj_col = []
for col in df.columns: 
    if(df[col].dtype == object):
        obj_col.append(col)
           
labelencoder = LabelEncoder()
for i in obj_col:
    df[i] = labelencoder.fit_transform(df[i].astype(str))   
print(df.info)
```

```python
# Features and Target variable
X=df.drop('isFraud',axis=1)
y=df['isFraud']
```

```python
# Using Pearson Correlation
plt.figure(figsize=(14, 12))  # Increase the figure size
cor = df.corr()

# Creating a heatmap with adjusted spacing and font scale
plt.subplots(figsize=(14, 12))
sns.set(font_scale=0.6)  # Adjust the font scale for annotation (try different values)
sns.heatmap(cor, annot=True, cmap="coolwarm", fmt=".2f")  # Use ".2f" to format annotations
plt.tight_layout()  # To avoid cropping the heatmap labels
plt.show()

# Correlation with output variable
cor_target = abs(cor["isFraud"])

# Selecting highly correlated features
relevant_features = cor_target[cor_target > 0]
print(relevant_features)

```

```python
from sklearn.model_selection import train_test_split
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

```

```python
# Print the shapes of the training and testing features and labels
print("Shape of training features (X_train):", X_train.shape)
print("Shape of training labels (y_train):", y_train.shape)
print("Shape of testing features (X_test):", X_test.shape)
print("Shape of testing labels (y_test):", y_test.shape)
```

Model Building
LOGisticRegression

```python
import timeit
start = timeit.default_timer()
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE
nr = SMOTE()
X_train, y_train = nr.fit_resample(X_train, y_train)
logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train, y_train)
y_pred=logreg.predict(X_test)
print(classification_report(y_test, y_pred))
stop = timeit.default_timer()
print('Time: ', stop - start)
```

```python
from sklearn.metrics import confusion_matrix, roc_auc_score, accuracy_score, precision_score, recall_score, roc_curve
errors = abs(y_pred - y_test)
print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')
cnf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cnf_matrix)
print("Accuracy:",accuracy_score(y_test, y_pred))
print("Precision:",precision_score(y_test, y_pred))
print("Recall:",recall_score(y_test, y_pred))
```

```python
y_pred_proba = logreg.predict_proba(X_test)[::,1]
fpr, tpr, _ = roc_curve(y_test,  y_pred_proba)
auc = roc_auc_score(y_test, y_pred_proba)
print(auc)
plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.title("ROC_AUC Curve")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.show()
```

```python
from keras import Sequential
from keras.layers import Dense
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
nr = SMOTE()
X_train, y_train = nr.fit_resample(X_train, y_train)
print('Training Features Shape:', X_train.shape)
print('Training Labels Shape:', y_train.shape)
print('Testing Features Shape:', X_test.shape)
print('Testing Labels Shape:', y_test.shape)
def build_model():    
    classifier = Sequential()
    classifier.add(Dense(6, activation='relu', kernel_initializer='random_normal', input_dim=28))
    classifier.add(Dense(6, activation='relu', kernel_initializer='random_normal'))
    classifier.add(Dense(1, activation='sigmoid', kernel_initializer='random_normal'))
    classifier.compile(optimizer ='adam',loss='binary_crossentropy', metrics =['accuracy'])
    return classifier
```

```python
start = timeit.default_timer()
keras_model = build_model()
keras_model.fit(X_train,y_train, batch_size=64, epochs=30)
print('Time: ', stop - start)
```
